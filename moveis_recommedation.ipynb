```python
def simplify(object):
    L = []
    for text in ast.literal_eval(object):
        if text['job'] == 'Director':
            L.append(text['name'])

    return L
```

```python
movies['crew'] = movies['crew'].apply(simplify)
```

```python
movies.head(10)
```

```python
movies['cast'][0]
```

```python
def ecxtract_cast(object):
    L = []
    counter = 0
    for text in ast.literal_eval(object):
        if counter < 3:
            L.append(text['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(ecxtract_cast)
```

```python
movies['cast']
```

```python
movies.head(10)
```

```python
movies['overview'][0]
```

```python
movies['overview']
```

```python
def convert(text):
    word = str(text).lower().split() 

    return word
```

```python
movies['overview'] = movies['overview'].apply(convert)
```

```python
movies['overview']
```

```python
movies['tagline'] = movies['tagline'].apply(convert)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['tagline']
```

```python
movies.head(10)
```

```python
movies.drop(['genres','keywords','overview','tagline','cast','crew'], axis= 1, inplace= True)
```

```python
movies.head(10)
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
movies.head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
Vector = cv.fit_transform(movies['tags']).toarray()
```

```python
Vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(Vector)
similarity
```

```python
sorted(similarity[0], reverse=True)
```

```python
index = movies[movies['title'] == 'Avatar']
index['id']
```

```python
def recommend(movie):
    index = movies[movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movies.iloc[i[0]].title)
        
```

```python
recommend('Spider-Man 3')
```
